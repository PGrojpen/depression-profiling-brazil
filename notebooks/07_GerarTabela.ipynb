```python
# Ler arquivo da base. Lembre de deixar a base e esse código na mesma pasta do seu computador para que a importação funcione.
db = pd.read_csv("database.csv")
```

```python
# Separe os atributos que deseja puxar de acordo com seu tipo
categorical = ["Q092", "P050", "P051", "P055", "P027","P02801", "P03201","P03001", "P03301", "P03302", "P03303"] 
continuous = ["Q09301", "P053", "P029", "P03202"]
```

```python
# Gerar tabela (caso apareça algum warning pode ignorar [contanto que a tabela seja criada de fato])
table1 = TableOne(db, dip_test=False, normal_test=False, tukey_test=False, categorical=categorical, continuous=continuous, min_max=continuous, include_null=True)
```

```python
# Mostrar tabela no formato latex
print(table1.tabulate(tablefmt = "latex"))
```

```python
# vamos corrigir a escrita da linha para dados do tipo continuo
for i in continuous:
    print(f"{i}, mean (SD / min / max) &      & {db[i].isnull().sum()}     & {db[i].mean():.1f} ({db[i].std():.1f} / {db[i].min()} / {db[i].max()}) \\\\")
```
