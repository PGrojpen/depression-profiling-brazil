# Análise de Entropia

## Objetivo
Realizar a análise de entropia para avaliar a relevância dos atributos presentes na base de dados, indentificando os atributos mais informativos

## Cálculo da Entropia dos Atributos

Para cada atributo da base de dados, foi calculada sua entropia individual utilizando a fórmula da Teoria da Informação:

\[
H(X) = -\sum_{i=1}^{n} p(x_i) \cdot \log_2 p(x_i)
\]

Onde:
- \( p(x_i) \) representa a probabilidade do valor \( x_i \) ocorrer no atributo.
- A entropia está expressa em *bits*, pois a base logarítmica utilizada é 2.

O cálculo foi realizado com base na **frequência relativa dos valores** de cada atributo. Valores ausentes foram removidos antes do cálculo para garantir a consistência da distribuição de probabilidade.

Em seguida, os atributos foram ranqueados em ordem decrescente de entropia, do mais informativo (maior entropia) ao menos informativo (menor entropia).


```python
# Dicionário para armazenar a entropia de cada coluna
entropyDict = {}

# Função que calcula entropia de uma coluna usando a biblioteca scipy
def calculate_entropy_using_scipy(column):
    counts = column.value_counts()
    probabilities = counts / counts.sum()
    return entropy(probabilities, base=2)
```

```python
# Cálculo da entropia para todas as colunas da base
for col in BasePoscaptopp.columns:
    entropy_value = calculate_entropy_using_scipy(BasePoscaptopp[col])
    entropyDict[col] = entropy_value
    print(f"Entropia da coluna {col}: {entropy_value}")
```

```python
# Ordenando os atributos por entropia (decrescente)
print("\n--- Entropias em ordem decrescente ---")
entropyDictSortedDecre = sorted(entropyDict, key=entropyDict.get, reverse=True)
for col in entropyDictSortedDecre:
    print(f"Entropia da coluna {col}: {entropyDict[col]}")

# Ordenando os atributos por entropia (crescente)
print("\n--- Entropias em ordem crescente ---")
entropyDictSortedCre = sorted(entropyDict, key=entropyDict.get)
for col in entropyDictSortedCre:
    print(f"Entropia da coluna {col}: {entropyDict[col]}")
```
